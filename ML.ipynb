```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(0)
x = 10* np.random.rand(100,1)
y = 2 + 2 * x + np.random.randn(100,1)
print(x)
print(y)
```

```python
#crea el modelo de  regrecion lineal 
model = LinearRegression()
#Entrenar el modelo con los datos de entrenamineto
model.fit(x, y)
```

```python
#Realizar predicciones con nuevos datos
x_new = np.array([[0], [10]])
y_pred = model.predict(x_new)
```

```python
#visualizar los resultados
plt.scatter(x, y)
plt.plot(x_new, y_pred, color = 'red')
plt.xlabel('x')
plt.ylabel('y')
plt.title("Regrecion Lineal")
plt.show
```
